# Extract collection

```python
soup=BeautifulSoup(driver.page_source, 'html.parser')
```

```python
results=soup.find_all('div', {'class': '_36IGc'})
```

```python
len(results)
```

```python
for i in range (0,1):
    item=results[i]
    
    description=item.h2.text
    title=description[5:]
    year=description.split()[0]
    
    price=item.h3.text[2:]
    
    kilometers=item.find('span').text[:-2]
    
    fuel_type=item.find('span',{'itemprop':'name'}).text
    
    owner=item.find_all('span')[3].text[:3]
    
print(title,year,kilometers,fuel_type,owner,price)
```

```python
def exctract(item):

    description=item.h2.text
    title=description[5:]
    year=description.split()[0]
    
    price=item.h3.text[2:]
    
    kilometers=item.find('span').text[:-2]
    
    fuel_type=item.find('span',{'itemprop':'name'}).text
    
    owner=item.find_all('span')[3].text[:3]
    
    result=(title,year,kilometers,fuel_type,owner,price)
    
    return result
```

```python
records=[]
for i in range(0,len(results)):
    info=results[i]
    record=exctract(info)
    records.append(record)    
```

```python
with open('results_kolkata.csv','w',newline='',encoding='utf-8') as f:
    writer=csv.writer(f)
    writer.writerow(['Name','Year','Kilometers_Driven','Fuel_Type','Owner_Type','Price in Lakhs'])
    writer.writerows(records)
```
